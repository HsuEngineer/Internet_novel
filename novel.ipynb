```python
from fpdf import FPDF
import os
import warnings

warnings.filterwarnings("ignore", message="cmap value too big/small")

def txt_to_pdf(txt_path, pdf_path, font_path="font/NotoSansTC-Regular.ttf"):
    pdf = FPDF()
    pdf.add_page()

    # ✅ 設定中文字型（此處用微軟正黑體，你也可以用其他 .ttf/.ttc）
    if not os.path.exists(font_path):
        raise FileNotFoundError(f"找不到字型：{font_path}")
    
    pdf.add_font("CJK", "", font_path, uni=True)
    pdf.set_font("CJK", size=12)

    # ✅ 寫入文字
    with open(txt_path, 'r', encoding='utf-8') as f:
        for line in f:
            pdf.multi_cell(0, 10, line.strip(), align='L')

    # ✅ 輸出 PDF
    pdf.output(pdf_path)
    print(f"✅ PDF 已建立：{pdf_path}")

os.chdir('F:/vscode/novel/')

#確保目錄存在
if not os.path.exists(f'content/{novel_name}/pdf'):
    os.mkdir(f'content/{novel_name}/pdf')
list = os.listdir(f'content/{novel_name}/txt')

#Example usage:
txt_file_path = f"content/{novel_name}/txt/{list[86]}"
pdf_file_path = f"content/{novel_name}/pdf/output.pdf"
txt_to_pdf(txt_file_path, pdf_file_path)
```
